```python
import requests
import pandas as pd

url = "https://fakestoreapi.com/products"
data = requests.get(url).json()

df = pd.DataFrame(data)

# Clean & structure
df = df[["id", "title", "price", "category", "rating"]]
df["rating_rate"] = df["rating"].apply(lambda x: x["rate"])
df["rating_count"] = df["rating"].apply(lambda x: x["count"])
df.drop("rating", axis=1, inplace=True)

df.rename(columns={
    "title": "Product_Name",
    "price": "Price",
    "category": "Category"
}, inplace=True)


big_df = pd.concat([df]*50, ignore_index=True)


import numpy as np

big_df["Price"] = big_df["Price"] + np.random.randint(-10, 20, size=len(big_df))
big_df["rating_rate"] = big_df["rating_rate"] + np.random.uniform(-0.5, 0.5, size=len(big_df))


big_df["rating_rate"] = big_df["rating_rate"].clip(1, 5)


big_df.to_csv("ecommerce_big_data.csv", index=False)

print("✅ Final Data Shape:", big_df.shape)
big_df.head()
```

```python
# importing reguired libraries 
import pandas as pd


# Loading data
df = pd.read_csv("ecommerce_big_data.csv")
df
```

```python
df.info()  # getting info about the dataset 

```

```python
df.dropna(inplace=True) # Remove rows with missing values to keep the dataset clean
print(df)


```

```python
df.drop_duplicates(inplace=True)  #droping duplicates here 
```

```python
# Convert price column into numeric format for calculations
df["Price"] = pd.to_numeric(df["Price"], errors="coerce")
df["rating_rate"] = pd.to_numeric(df["rating_rate"], errors="coerce")
df["rating_count"] = pd.to_numeric(df["rating_count"], errors="coerce")
```

```python
df = df[df["Price"] > 0]
df = df[df["rating_rate"] >= 1]
df = df[df["rating_rate"] <= 5]
```

```python
df["Price_Category"] = pd.cut(
    df["Price"],
    bins=[0, 50, 150, 500, 1000],
    labels=["Low", "Medium", "High", "Premium"]
)
```

```python


df["Popularity"] = pd.cut(
    df["rating_count"],
    bins=[0, 100, 300, 600, 1000],
    labels=["Low", "Medium", "High", "Very High"]
)

```

```python
df["Product_Name"] = df["Product_Name"].str.strip()
```

```python
df.to_csv("cleaned_ecommerce_data.csv", index=False)
```
